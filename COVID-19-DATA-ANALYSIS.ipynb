```python
import ipywidgets as widgets
```

```python
x1 = widgets.Dropdown(
    options=df1.columns,
    value=df.columns[0],
    description='X_val:',
    disabled=False,
    )

def xtake(val1):
    print('first Function', val1)
    
    x2 = widgets.Dropdown(
    options=['lineplot','scatterplot'],
    value='lineplot',
    description='Graphs:',
    disabled=False,
    )
    
    def graphviz2(val2):
        print('Second Function',val2)
        
        if val2 == 'lineplot':
            sns.lineplot(x = val1, y = 'Cured', data = df1)
        
        if val2 == 'scatterplot':
            sns.scatterplot(x = val1, y = 'Cured', data = df1)
            
    widgets.interact(graphviz2,val2=x2)
    
```

```python
widgets.interact(xtake,val1=x1)            # Importing widgets and taking variables for interactive graphs
```
